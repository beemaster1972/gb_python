```python
import pandas as pd
import random
```

```python

lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
one_hot_human = [1 if who_am_i == 'human' else 0 for who_am_i in data['whoAmI']]
one_hot_robot = [1 if who_am_i == 'robot' else 0 for who_am_i in data['whoAmI']]
data = data.copy()
data.assign(human = one_hot_human)
data.assign(robot = one_hot_robot)
data
```
